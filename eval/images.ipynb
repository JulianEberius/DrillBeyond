```python
def to_relation_fragment(y,q):
    name, sel = y.name.split("_")[:2]
    y.index.name="k"
    y = y.reset_index()
    name, sel = y.columns[1].split("_")[:2]

    y['method'] = len(y) * [name]
    y['sel'] = len(y) * [sel]
    y['Q'] = len(y) * [q]
    y = y.rename(columns={y.columns[1]:"time"})
    return y

def universal_df(S):
    df = pd.concat([to_relation_fragment(v,k) for k,lst in S.iteritems() for v in lst], ignore_index=True).convert_objects(convert_numeric=True)
    df = df.set_index(["Q","k","method","sel"]).sortlevel().unstack(2)
    return df

def diffToMin(Z,algos):
    n = len(algos)
    Z = Z.copy()
    Z['min'] = Z.apply(lambda row: min(row), axis=1)

    for c in algos:
        Z[c+'Diff'] = Z[c] - Z['min'] + 0.00001

    return Z.iloc[:,n+1:]

def percentOfMax(Z):
    algos = Z.columns
    n = len(algos)
    Z = Z.copy()
    Z['max'] = Z.apply(lambda row: max(row), axis=1)

    for c in algos:
        Z[c+'Norm'] = Z[c] / Z['max']

    return Z.iloc[:,n+1:]

def percentOfMin(Z):
    algos = Z.columns
    n = len(algos)
    Z = Z.copy()
    Z['min'] = Z.apply(lambda row: min(row), axis=1)

    for c in algos:
        Z[c+'Norm'] = Z[c] / Z['min']

    Z = Z.iloc[:,n+1:]
    Z.rename(columns=lambda c: c.replace("Norm","").replace("DefS",""), inplace=True)
    Z.rename(columns={"bigOmegaDynamicStaticReOpt":"bigOmegaDynamicReOpt"}, inplace=True)
    return Z
```

```python
#db_file_name = "test_drb.db"
db_file_name = "corr.db"
conn = sqlite3.connect("%s" % db_file_name)
q = """select
    concept,
    case
        when attribute = 'revenues|revenue|sales' then 'revenues'
        when attribute = 'revenues growth|revenue growth|sales growth' then 'revenueGrowth'
        when attribute = 'profit|earnings|profits' then 'profit'
        when attribute = 'established|founded' then 'founded'
        when attribute = 'population|inhabitants' then 'population'
        when attribute = 'size|area' then 'area'
        when attribute = 'revenues growth|revenue growth|sales growth|revenues change|revenues change|sales change' then 'revenueGrowth'
        when attribute = 'population growth' then 'populationGrowth'
        else attribute
    end as attribute,
    case
        when predicates = '' then 'noPredicate'
        when predicates = 'isNumeric' then 'isNumeric'
        when predicates like '%|%' then 'bothPredicates'
        when predicates like '>%' then 'numericPredicate'
        when predicates like '<%' then 'numericPredicate'
        else predicates
    end as predicate,
    case
        when inputFile like '%andom%' then 'random'
        else 'top'
    end as type,
    case
        when method = 'GreedySetCoverer' then 'Greedy'
        when method = 'GeneticSetCovererQ' then 'Genetic'
    end as method,
    cp.entity, cp.rank, a.correct
from
    experiments e
    left join coverParts cp on e.id=cp.id
    left join answers a on cp.identifier = a.identifier and cp.entity=a.entity and e.attribute=a.forAttribute
where
    inputFile not like '%andom%b%'
--    method='GeneticSetCovererQ'
;"""
precs = pd.read_sql(q, conn)
```
